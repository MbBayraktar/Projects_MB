## <span style=" color:red">Determines 
    
Features
    
- timestamp - timestamp field for grouping the data
- cnt - the count of a new bike shares
- t1 - real temperature in C
- t2 - temperature in C feels like
- hum - humidity in percentage
- wind_speed - wind speed in km/h
- weather_code - category of the weather
- is_holiday - boolean field - 1 holiday / 0 non holiday
- is_weekend - boolean field - 1 if the day is weekend
- season - category field meteorological seasons: 0-spring ; 1-summer; 2-fall; 3-winter.
    
"weather_code" category description:

- 1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity
- 2 = scattered clouds / few clouds
- 3 = Broken clouds
- 4 = Cloudy
- 7 = Rain/ light Rain shower/ Light rain
- 10 = rain with thunderstorm
- 26 = snowfall
- 94 = Freezing Fog

# <span style=" color:red">Conclusions
    

    

</span>

**1.** <span style="color:blue">The data set includes regular hourly measurements of the bicycle sharing system used in urban transportation in London, covering the period between 04.01.2015 and 03.01.2017.</span>

**2.** <span style="color:blue">During the two-year period, the system was used a total of 20 million times, an average of 27,543 times per day and 1,143 times per hour.</span>

**3.** <span style="color:blue">There is a moderate negative relationship between the humidity of the air and the use of the system. There is also a moderate positive relationship with Temperature. There is a slight positive relationship with wind speed.</span>

**4.** <span style="color:blue">It reached the highest hourly usage figures on 9 July 2015 and 06 August 2015. In the research conducted, it was determined that there was a general public transportation strike in the city on these dates.</span>

**5.** <span style="color:blue">Average bicycle use reaches its peak in July each year, then enters a downward trend and an upward trend in January.</span>

**6.** <span style="color:blue">The system was used most between 08:00 and 17:00 and least between 04:00.</span>

**7.** <span style="color:blue">Although the number of uses during the daytime during holidays does not show a significant difference, it has been observed that the peak usage is reached around 14:00.</span>

**8.** <span style="color:blue">The order of cycling based on season, from most to least, is Summer, Autumn, Spring, Winter.</span>

**9.** <span style="color:blue">Average daily usage has always been higher on non-holiday weekdays. When looking at weekends, there is slightly less usage.</span>

**10.** <span style="color:blue">Although not regular, the usage trend shows a downward trend from the beginning of the month to the end of the month. It can also be seen that the average daily usage has the lowest value on the 31st day of the month.</span>
