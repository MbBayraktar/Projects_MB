# How Does the Funding Ecosystem changes with respect to Time?

```python
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
funding_by_year.plot(kind='line', marker='o', color='skyblue')
plt.title('Number of Funding by Year')
plt.xlabel('Year')
plt.ylabel('Number of Funding')

plt.subplot(3, 1, 2)
funding_by_quarter.plot(kind='line', marker='s', color='orange') 
plt.title('Number of Funding by Quarter')
plt.xlabel('Quarter')
plt.ylabel('Number of Funding')

plt.subplot(3, 1, 3)
funding_by_month.plot(kind='line', marker='d', color='green') 
plt.title('Number of Funding  by Month')
plt.xlabel('Month')
plt.ylabel('Number of Funding')

plt.tight_layout()
plt.show()
```

# Conclusions

**1.** <span style="color:blue">Over an approximate 5-year period, a total of 2,113 investors made 2,157 investments in 590 sectors across 106 cities in India, using 16 different investment methods, with a total investment amount of $43 billion USD.</span>

**2.** <span style="color:blue">The average annual number of investments during this five-year period is 435, with the highest number of investments at 989 in 2016 and the lowest at 110 in 2019. (except : 2020)</span>

**3.** <span style="color:blue">The average monthly number of investments over the five-year period is 218, with the highest number of investments totaling 266 in June and the lowest totaling 186 in December.</span>

**4.** <span style="color:blue">The average number of investments per quarter over the five-year period is 653, with the highest number of investments totaling 733 in the 1st quarter and the lowest totaling 603 in the 4th quarter.</span>

**5.** <span style="color:blue">Over a five-year period, the average investment amount is 16.5 million USD, 
with the highest investment amount being 3.9 billion USD made by Westbridge Capital in the transportation sector for the Rapido Bike Taxi company in Bengaluru. The lowest investment amount is 18 thousand USD, made by The Ten Minute Million in the consumer internet sector for the Maptags company in Bengaluru.</span>

**6.** <span style="color:blue">The top 10 startup firms have included into the consumer internet sector with 935 startups,
the technology sector with 477 startups and the e-commerce sector with 287 startups, respectively. These startups account for 92% of the total number of startups in the top 10.</span>
    
**7.** <span style="color:blue">The top 3 cities with the highest number of investments are Bengaluru with 778 entries, Mumbai with 516 entries, and New Delhi with 397 entries.</span>
    
**8.** <span style="color:blue">The majority of investors have not disclosed their names. Among other investors, Indian Angel Network is the company that has made the highest number of investments in terms of quantity. However, there is no significant difference observed among the other investors.</span>
    
**9.** <span style="color:blue">The total number of investment types used by investors is 16, and the two most preferred types by investors are Seed and Private, respectively.</span>
